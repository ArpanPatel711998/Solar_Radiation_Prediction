# Data cleaning

```python
data['SunRise_hour'] = pd.to_datetime(data['TimeSunRise']).dt.hour
data['SunRise_minute'] = pd.to_datetime(data['TimeSunRise']).dt.minute
data
```

```python
data.drop(data.columns[7],axis = 1,inplace = True)
data
```

```python
data['SunSet_hour'] = pd.to_datetime(data['TimeSunSet']).dt.hour
data['SunSet_minute'] = pd.to_datetime(data['TimeSunSet']).dt.minute
data
```

```python
data.drop(data.columns[7],axis = 1,inplace = True)
data
```

```python
import datetime


x = []
for i in data['UNIXTime']:
    date_time = datetime.datetime.fromtimestamp(i)
    x.append(date_time.strftime('%Y-%m-%d %H:%M:%S'))

x
```

```python
s = {'UNIXTime_convet':pd.Series(x)}
s = pd.DataFrame(s)
s
```

```python
data = pd.concat([data,s],axis = 1)
data
```

```python
data['UNIXTime_day'] = data['UNIXTime_convet'].dt.day
data['UNIXTime_month'] = data['UNIXTime_convet'].dt.month
data['UNIXTime_year'] = data['UNIXTime_convet'].dt.year
data['UNIXTime_hour'] = data['UNIXTime_convet'].dt.hour
data['UNIXTime_minute'] = data['UNIXTime_convet'].dt.minute
data
```

```python
data = data.drop(columns = ['UNIXTime_convet'], axis = 1)
data
```

# Outlier Remove

```python
sns.boxplot(x = data['Radiation'])
```

```python
percentile25 = data['Radiation'].quantile(0.25)
percentile75 = data['Radiation'].quantile(0.75)
```

```python
IQR = percentile75 - percentile25
```

```python
upper_limit = percentile75 + 1.5*IQR
lower_limit = percentile25 - 1.5*IQR
```

```python
data[data['Radiation'] > upper_limit]
```

```python
data[data['Radiation'] < lower_limit]
```

```python
new_df = data[data['Radiation'] < upper_limit] 
```

```python
data = pd.concat([new_df],ignore_index=True)
data
```

```python
data['Radiation'].plot()
```

# test_data cleaning

```python

test_data['SunRise_hour'] = pd.to_datetime(test_data['TimeSunRise']).dt.hour
test_data['SunRise_minute'] = pd.to_datetime(test_data['TimeSunRise']).dt.minute

test_data = test_data.drop(columns = 'TimeSunRise',axis=1)






test_data['SunSet_hour'] = pd.to_datetime(test_data['TimeSunSet']).dt.hour
test_data['SunSet_minute'] = pd.to_datetime(test_data['TimeSunSet']).dt.minute

test_data = test_data.drop(columns = 'TimeSunSet',axis=1)






x = []
for i in test_data['UNIXTime']:
    date_time = datetime.datetime.fromtimestamp(i)
    x.append(date_time.strftime('%Y-%m-%d %H:%M:%S'))
    
    
    

s = {'UNIXTime_convet':pd.Series(x)}
s = pd.DataFrame(s)



test_data = pd.concat([test_data,s],axis = 1)





test_data['UNIXTime_convet'] = pd.to_datetime(test_data['UNIXTime_convet'])




test_data['UNIXTime_day'] = test_data['UNIXTime_convet'].dt.day
test_data['UNIXTime_month'] = test_data['UNIXTime_convet'].dt.month
test_data['UNIXTime_year'] = test_data['UNIXTime_convet'].dt.year
test_data['UNIXTime_hour'] = test_data['UNIXTime_convet'].dt.hour
test_data['UNIXTime_minute'] = test_data['UNIXTime_convet'].dt.minute




test_data = test_data.drop(columns = ['UNIXTime_convet','UNIXTime'], axis = 1)
test_data
```

# test_data outlier remove

```python
percentile25 = test_data['Radiation'].quantile(0.25)
percentile75 = test_data['Radiation'].quantile(0.75)



IQR = percentile75 - percentile25


upper_limit = percentile75 + 1.5*IQR
lower_limit = percentile25 - 1.5*IQR


test_data[test_data['Radiation'] > upper_limit]


test_data[test_data['Radiation'] < lower_limit]


test_data = test_data[test_data['Radiation'] < upper_limit] 


test_data = pd.concat([test_data],ignore_index=True)
test_data
```
